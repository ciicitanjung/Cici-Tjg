```python
# latihan 3
start = int(input("Masukkan angka awal: "))
end = int(input("Masukkan angka akhir: "))

# Fungsi untuk mengecek apakah suatu bilangan adalah bilangan prima
def prima(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

num = start

while num <= end:
    if num % 3 == 0 or num % 5 == 0:
        num += 1
        continue

    if prima(num):
        print(f"Bilangan prima pertama dalam rentang {start}-{end} yang tidak habis dibagi 3 atau 5 adalah: {num}")
        break

    num += 1
else:
    print("Tidak ada bilangan prima yang memenuhi syarat dalam rentang tersebut.")
```

```python
# Tugas 2
# Program untuk mencari semua solusi bilangan bulat tidak negatif dari x + y + z = 25

total = 25

for x in range(total + 1):
    for y in range(total - x + 1):
        z = total - x - y
        print(f"(x={x}, y={y}, z={z})")
```

```python
# Tugas 3
saldo = 50000  # Saldo awal
saldo_minimum = 10000  # Saldo minimum yang harus tersisa

while True:
    print(f"Saldo saat ini: Rp{saldo}")
    kode = int(input("Masukkan kode transaksi (0: Setor, 1: Tarik, 2: Keluar): "))

    if kode == 2:
        print("Terima kasih! Program selesai.")
        break
    elif kode not in [0, 1]:
        print("Kode transaksi tidak valid. Gunakan 0 untuk setor atau 1 untuk tarik.")
        continue

    jumlah = int(input("Masukkan jumlah uang: "))

    if kode == 0:  # Setor uang
        saldo += jumlah
    elif kode == 1:  # Tarik uang
        if saldo - jumlah < saldo_minimum:
            print("Transaksi gagal! Saldo tidak boleh kurang dari Rp10.000.")
        else:
            saldo -= jumlah

    if saldo == saldo_minimum:
        print("Saldo minimum tercapai. Tidak dapat melakukan transaksi lagi.")
        break
```
